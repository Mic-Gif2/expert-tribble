```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Construction_Data_PM_Tasks_All_Projects.csv')

# Group by 'Task Group' and count the occurrences of 'Open' and 'Closed' in 'Report Status'
status_counts = df.groupby('Task Group')['Report Status'].value_counts().unstack().fillna(0)

# Plotting
status_counts.plot(kind='bar', stacked=False, figsize=(10,7))
plt.title("Number of Open and Closed Tasks by Task Group")
plt.ylabel("Number of Tasks")
plt.xlabel("Task Group")
plt.tight_layout()
plt.legend(title="Report Status")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Construction_Data_PM_Tasks_All_Projects.csv')

# Filter rows where "OverDue" is True and then group by the project column to count occurrences
overdue_counts = df[df['OverDue'] == True].groupby('project').size()

# Plotting
overdue_counts.sort_values(ascending=False).plot(kind='bar', figsize=(12,8))
plt.title("Number of Overdue Tasks by Project")
plt.ylabel("Number of Overdue Tasks")
plt.xlabel("Project")
plt.tight_layout()
plt.show()
```
